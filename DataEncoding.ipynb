```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

#Create a simple dataframe

df = pd.DataFrame({
    'color' : ['red', 'blue', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
#Create an instance of OHE

encoder = OneHotEncoder()

#Perform fit then transform

encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
#For new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df, encoded_df], axis = 1)
```

```python
#Ordinal Encoding

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({
    'size' :  ['small', 'medium', 'large', 'medium', 'large']
})

df.head()
```

```python
ordEnc = OrdinalEncoder(categories=[['small', 'medium', 'large']])

ordEnc.fit_transform(df[['size']])
```

```python
#Target Guided Ordinal Encoding
import pandas as pd
df = pd.DataFrame({
    'city' : ['new york', 'tokyo', 'mumbai', 'japan', 'new york', 'mumbai'],
    'price' : [180, 200, 140, 300, 250, 700]
})

df
```

```python

mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
df['encoded_price'] = df['city'].map(mean_price)

df
```
